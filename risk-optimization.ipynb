```python
import scipy.optimize
from scipy.optimize import minimize
```

```python
def F(x, q1, q2, q3, q4, q5, q6, q7, q8, q9):
    r1, r2, r3, r4, r5, r6, r7, r8, r9 = x
    return q1 * r1 + q2 * r2 + q3 * r3 + q4 * r4 + q5 * r5 + q6 * r6 + q7 * r7 + q8 * r8 + q9 * r9

def constraint1(x):
    r1, r2, r3, r4, r5, r6, r7, r8, r9 = x
    return r1, r2, r3, r4, r5, r6, r7, r8, r9

def constraint2(x, q1, q2, q3, q4, q5, q6, q7, q8, q9):
    return 1459317 - (q1 * x[0] + q2 * x[1] + q3 * x[2] + q4 * x[3] + q5 * x[4] + q6 * x[5] + q7 * x[6] + q8 * x[7] + q9 * x[8])
```

```python
q1 = 31200
q2 = 22550.64
q3 = 51200
q4 = 42282
q5 = 65000
q6 = 89564
q7 = 6000
q8 = 24664.5
q9 = 4000

x0 = [0.12, 0.18, 0.16, 0.15, 0.1, 0.1, 0.1, 0.05, 0.04]
```

```python
bound = scipy.optimize.Bounds(lb=0.001, ub=1)

cons = [{'type': 'ineq', 'fun': constraint1}, {'type': 'ineq', 'fun': constraint2, 'args': (q1, q2, q3, q4, q5, q6, q7, q8, q9)}]
```

```python
res = minimize(F, x0, args=(q1, q2, q3, q4, q5, q6, q7, q8, q9), bounds=bound, constraints=cons)
res.x
```
